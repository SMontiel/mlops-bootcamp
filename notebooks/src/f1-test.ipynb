```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
circuit_cols = [
    'circuitId', 'circuitRef', 'name', 'location', 'country', 'lat', 'lng', 'alt', 'url'
]
circuits = pd.read_csv('datasets/circuits.csv', names=circuit_cols, low_memory=False, header=0)

races_cols = [
    'raceId', 'year', 'round', 'circuitId', 'name', 'date', 'time', 'url', 'fp1_date', 'fp1_time', 'fp2_date',
    'fp2_time', 'fp3_date', 'fp3_time', 'quali_date', 'quali_time', 'sprint_date', 'sprint_time']
races = pd.read_csv('datasets/races.csv', names=races_cols, low_memory=False, header=0)

#races.info()
races.head(5)
```

```python
# Drop columns
circuits.drop(['lat', 'lng', 'alt', "url"], axis = 'columns', inplace = True)

races.drop(["url", 'time'], axis = 'columns', inplace = True)

#races.info()
races.head(5)

races.sort_values(by=['year', 'raceId', 'round'], inplace=True, ascending=False)
races.head(70)
```
